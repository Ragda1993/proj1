```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df = pd.read_csv(r"/content/Sleep_health_and_lifestyle_dataset.csv")
print (df)
```

```python
df.dtypes
```

```python
# changes the type of Sleep Duration to string
df["Sleep Duration"] = df["Sleep Duration"].astype("str")
```

```python
df.info()
```

```python
# remove extra spaces from each entry
df["Sleep Duration"].str.strip(" ")
```

```python
# return the type of Sleep Duration to the float
df["Sleep Duration"] = df["Sleep Duration"].astype("float")
```

```python
df['BMI Category'].replace('Normal Weight ','Normal',inplace = True)
```

```python
df["BMI Category"]
```

```python
# split data by gender
filt = (df["Gender"] == 'Female')
df.loc[filt]
```

```python
df.dtypes
```

```python
# How many females suffer from insomnia ?
sum(df.loc[filt]['Sleep Disorder'] == "Insomnia")
```

```python
# How many females suffer from sleep apnea ?
sum(df.loc[filt]['Sleep Disorder'] == "Sleep Apnea")
```

```python
filt1  = (df["Gender"] == 'Male')
df.loc[filt1]
```

```python
#  How many males suffer from insomnia ?
sum(df.loc[filt1]['Sleep Disorder'] == "Insomnia")
```

```python
#  How many females suffer from sleep apnea ?
sum(df.loc[filt1]['Sleep Disorder'] == "Sleep Apnea")
```

# conclusion:
We notice that into two cases the number of Females who suffer from sleep apnea is greater than from
isomnia but in the males it is the opposite .

```python
# What is the percentage of gender in each occpation ?
df['Occupation'].value_counts(normalize=True)*100
```

```python
# Draw a diagram showing the percentage of Gender in each Occpation ?
df['Occupation'].value_counts(normalize=True).plot(kind='bar')
```

```python
# count the level of Stress level ?
df['Stress Level'].value_counts()
```

```python
# what the number of Occupation based on BMI Category ?
df[["Occupation","BMI Category"]].value_counts(normalize=True)*100
```

```python
# Draw a diagram showing the division of Occupation based on BMI Category?
ax=df[["Occupation","BMI Category"]].value_counts().plot(kind='barh')
ax.set_xlabel('BMI Category')
ax.set_ylabel('Occupation')
```

```python
# what the number of Occupation based on Stress Level ?
df [["Occupation","Stress Level"]].value_counts()
```

```python
# Draw a diagram showing the division of Occupation based on Stress Level ?
ax = df [["Occupation","Stress Level"]].value_counts() .plot(kind='barh')
ax.set_xlabel(' Occupation')
ax.set_ylabel('Stress Level')
```

```python
# Compare Age , Quality of Sleep ,Sleep Duration against each other ?
sns.pairplot(df,
             vars=["Age", "Quality of Sleep","Sleep Duration"])
plt.show()
```

```python
# Show the correlation between them ?
df_corr = df[["Age", "Quality of Sleep","Sleep Duration"]].corr()
df_corr
```

```python
sns.heatmap(df_corr , annot=True)
```

# Conclusion :
From the above there is strong relationship between Sleep Duration and Quality of Sleep .

```python
# Compare Physical Activity Level , Quality of Sleep , Daily Steps against each other ?
sns.pairplot(df,
             vars=["Physical Activity Level", "Quality of Sleep","Daily Steps"])
plt.show()
```

```python
# Show correlation between them ?
df_corr1 = df[["Physical Activity Level", "Quality of Sleep","Daily Steps"]].corr()
df_corr1
```

```python
sns.heatmap(df_corr1 , annot=True)
```

# conclusion:
From the above we notice that there is strong relationship between Daily Steps
 and Physical activity level ,and the quality of sleep does not affect either of them .
